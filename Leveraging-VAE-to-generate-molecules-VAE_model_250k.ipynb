```
data_path = "/content/drive/My Drive/zinc/250k_smiles.csv"

import csv
import numpy as np

import pandas as pd

#read file 
df = pd.read_csv(data_path)

#dropping the other 3 columns which includes the properties of the molecules
to_drop = ['logP', 'qed', 'SAS']

df.drop(to_drop, inplace=True, axis=1)
           
```

```
#print(df.head()) 
data_smiles = df.to_numpy()
```

```
data_array = []
for i in range(0, len(data_smiles)):
  data_array.append(data_smiles[i][0])

import numpy as np
train_smiles = np.asarray(data_array) 
print(type(train_smiles))

print(train_smiles[0])
```

```
tokens = set()
for s in train_smiles:
  tokens = tokens.union(set(s))
tokens = sorted(list(tokens))
max_length = max(len(s) for s in train_smiles)
```

```
import tensorflow as tf
import deepchem as dc

#the ecoder = CNN and the decoder = GRU
from deepchem.models.tensorgraph.optimizers import Adam, ExponentialDecay
from deepchem.models.tensorgraph.models.seqtoseq import AspuruGuzikAutoEncoder
model = AspuruGuzikAutoEncoder(tokens, max_length, model_dir='vae')

batches_per_epoch = len(train_smiles)/model.batch_size
learning_rate = ExponentialDecay(0.001, 0.95, batches_per_epoch)
model.set_optimizer(Adam(learning_rate=learning_rate))
```

```
def generate_sequences(epochs): 
  for i in range(epochs):
    for s in train_smiles: 
      yield (s, s)

#deepchem has its own fit model variation
model.fit_sequences(generate_sequences(10))
```

```
import numpy as np
from rdkit import Chem
predictions = model.predict_from_embeddings(np.random.normal(size=(1000,196))) 
molecules = []

#using chem from rdkit to ensure generated molecules are valid
for p in predictions:
  smiles = ''.join(p)
  if Chem.MolFromSmiles(smiles) is not None:
    molecules.append(smiles) 

for m in molecules:
  print(m)
```

```
smiles_list = [Chem.MolFromSmiles(x) for x in molecules]
```

```
print(sorted([x.GetNumAtoms() for x in smiles_list]))
```

```
good_mol_list = [x for x in smiles_list if x.GetNumAtoms() > 10 
                 and x.GetNumAtoms() < 50]
#printing number of good generated molecules
print(len(good_mol_list))
```

```
#obtaining QED and dropping molecules with QED below 0.5
from rdkit.Chem import QED
qed_list = [] 
for x in good_mol_list:
  try:
    qed_list.append(QED.qed(x))
  except ValueError:
    pass
final_mol_list = [(a,b) for a,b in
  zip(good_mol_list,qed_list) if b > 0.5]
```

```
#printing the molecules in the final filtered list of generated compounds
for i in final_mol_list:
  print(i)
```

```
from rdkit.Chem.Draw import MolsToGridImage

#outputing structures of molecules
MolsToGridImage([x[0] for x in final_mol_list], molsPerRow=3,useSVG=True, 
                subImgSize=(250, 250),
                legends=[f"{x[1]:.2f}" for x in final_mol_list])
```
